#Searching Algorithms
What is Searching?

Searching is the process of finding whether a particular element exists in a collection of data such as an array, list, string, or tree.

Example

Suppose we have:

Index : 0  1  2  3  4  5

Array : 12 18 25 40 56 90

Target = 40

Searching means finding:

Is 40 present?
If yes, at which index?

Output:

40 found at index 3

| Searching Algorithm  | Array Required                | Time Complexity      |
| -------------------- | ----------------------------- | -------------------- |
| Linear Search        | Sorted or Unsorted            | O(n)                 |
| Binary Search        | Sorted Only                   | O(log n)             |
| Jump Search          | Sorted                        | O(√n)                |
| Interpolation Search | Sorted (Uniform Distribution) | O(log log n) Average |
| Exponential Search   | Sorted                        | O(log n)             |
| Fibonacci Search     | Sorted                        | O(log n)             |
| Ternary Search       | Sorted                        | O(log₃ n)            |
| Hash Search          | Hash Table                    | O(1) Average         |
| DFS Search           | Graph/Tree                    | O(V + E)             |
| BFS Search           | Graph/Tree                    | O(V + E)             |


#1. Linear Search
Definition

Linear Search checks every element one by one until the target is found or the array ends.

No sorting is required.


#Working

Array

10 20 30 40 50

Target = 40

Step 1

10 ❌

Step 2

20 ❌

Step 3

30 ❌

Step 4

40 ✅ Found
Algorithm
Start

↓

Compare first element

↓

Found?

↓

Yes → Stop

↓

No

↓

Move to next element

↓

Repeat

↓

End of array?

↓

Not Found

#Time Complexity

Best Case

O(1)

Target is first element.

Average Case

O(n)

Worst Case

O(n)

Target is last element or absent.

Space Complexity
O(1)

#Advantages
Very easy to implement
,Works on unsorted arrays
,No preprocessing needed
#Disadvantages
Slow for large datasets
,Checks elements one by one

#Applications
Small datasets
Unsorted arrays
Linked Lists

#Program 1(Linear Seach)

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            print("Array Found bro")
            return i   # Return the index

    print("Array Not Found bro")
    return -1

result = linear_search([10, 20, 30, 40, 50], 40)
print(result)
```

#Question1:
list of study grades, count how many students scored above 85

```python
def count_grades(grades):
  count=0
  for grade in grades:
    if grade>=85:
      count +=1
  return count
print(count_grades([85,90,95,75,84,91]))
```

#Question 2:. Count Multiples of 5

```python
def multiples_of_five(nums):

    count = 0

    for num in nums:
        if num % 5 == 0:
            count += 1

    return count

print(multiples_of_five([10,15,22,30,41]))
```
