## Merge Sort Algorithm

Merge Sort is an efficient, comparison-based, divide and conquer sorting algorithm. Most implementations produce a stable sort, meaning that the relative order of equal sort items is preserved.

**How it works:**

1.  **Divide:** The unsorted list is divided into `n` sublists, each containing one element (a list of one element is considered sorted).
2.  **Conquer (Recursion):** Repeatedly merge sublists to produce new sorted sublists until there is only one sublist remaining. This will be the sorted list.

**Detailed Steps:**

Let's say we have an array `arr`.

1.  **Splitting (Divide):**
    *   The `merge_sort` function takes an array `arr` as input.
    *   It checks if the length of `arr` is greater than 1. If it's not, the array is already sorted (base case for recursion).
    *   It finds the middle index of the array (`mid = len(arr) // 2`).
    *   It splits the array into two halves: `left_half = arr[:mid]` and `right_half = arr[mid:]`.
    *   It recursively calls `merge_sort` on `left_half` and `right_half`. This continues until each sublist contains only one element.

2.  **Merging (Conquer):**
    *   Once the arrays are split down to single elements, the `merge` process begins.
    *   The `merge` operation takes two sorted sublists and merges them into a single sorted list.
    *   Three pointers (`i`, `j`, `k`) are used:
        *   `i` for iterating through `left_half`.
        *   `j` for iterating through `right_half`.
        *   `k` for iterating through the main `arr` (where the merged result will be stored).
    *   While `i` is less than the length of `left_half` AND `j` is less than the length of `right_half`:
        *   Compare `left_half[i]` and `right_half[j]`.
        *   If `left_half[i]` is smaller or equal, place `left_half[i]` into `arr[k]` and increment `i`.
        *   Else, place `right_half[j]` into `arr[k]` and increment `j`.
        *   In both cases, increment `k`.
    *   After one of the sublists is exhausted, there might be remaining elements in the other sublist. Copy these remaining elements directly into `arr`.

**Time Complexity:** O(n log n) in all cases (best, average, worst).
**Space Complexity:** O(n) due to the temporary arrays created during the merge process.

Merge Sort is often preferred for sorting linked lists because it performs well with them, and for external sorting due to its efficient handling of data that doesn't fit into memory.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Find the middle of the array
        left_half = arr[:mid]  # Divide the array elements into two halves
        right_half = arr[mid:]

        merge_sort(left_half)  # Sort the first half
        merge_sort(right_half)  # Sort the second half

        i = j = k = 0

        # Copy data to temp arrays left_half[] and right_half[]
        while i < len(left_half) and j < len(right_half):
            if left_half[i] <= right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Checking if any elements were left
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

# Demo of Merge Sort
if __name__ == '__main__':
    data = [12, 11, 13, 5, 6, 7]
    print("Original array:", data)
    merge_sort(data)
    print("Sorted array:", data)

    data2 = [38, 27, 43, 3, 9, 82, 10]
    print("Original array:", data2)
    merge_sort(data2)
    print("Sorted array:", data2)
```

```python
def merge_sort(arr):

    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])

    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):

    result = []

    i = 0
    j = 0

    while i < len(left) and j < len(right):

        if left[i] <= right[j]:
            result.append(left[i])
            i += 1

        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])

    result.extend(right[j:])

    return result


numbers = [38,27,43,3,9,82,10]

sorted_numbers = merge_sort(numbers)

print(sorted_numbers)
```
