#Pattern Programs


```python
rows=5

for i in range(1,rows+1):
  for j in range(i):
    print("*",end=" ")
  print("")
```

```python
rows=5

for i in range(rows,0,-1):
  for j in range(i):
    print("*",end=" ")
  print("")
```

```python
rows=5

for i in range(1,rows+1):
  for j in range(1,i+1):
    print(j,end=" ")
  print("")
```

```python
rows=5

for i in range(1,rows+1):
  for j in range(i):
    print(chr(65 + j),end=" ")
  print("")
```

```python
rows = 5

for i in range(1, rows + 1):

    # Print spaces
    for j in range(rows - i):
        print(" ", end="")

    # Print stars
    for k in range(2 * i - 1):
        print("*", end="")

    print()
```

```python
rows = int(input("Enter the number of rows: "))

for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```

```python
rows = 5

for i in range(rows, 0, -1):

    # Print spaces
    for j in range(rows - i):
        print(" ", end="")

    # Print stars
    for k in range(2 * i - 1):
        print("*", end="")

    print()
```

#Matrix Addition

```python
# Matrix Addition

rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

A = []
B = []

print("\nEnter elements of Matrix A:")
for i in range(rows):
    row = []
    for j in range(cols):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)


print("\nEnter elements of Matrix B:")
for i in range(rows):
    row = []
    for j in range(cols):
        row.append(int(input(f"B[{i}][{j}] = ")))
    B.append(row)

result = []

for i in range(rows):
    row = []
    for j in range(cols):
        row.append(A[i][j] + B[i][j])
    result.append(row)

print("\nResult Matrix:")
for row in result:
    print(row)
```

#Matrix Multiplication

```python
# Matrix Multiplication

rows = 2
cols = 2

A = []
B = []

print("Enter Matrix A")

for i in range(rows):
    row = []
    for j in range(cols):
        row.append(int(input()))
    A.append(row)

print("Enter Matrix B")

for i in range(rows):
    row = []
    for j in range(cols):
        row.append(int(input()))
    B.append(row)

result = [[0,0],[0,0]]

for i in range(rows):
    for j in range(cols):
        for k in range(cols):
            result[i][j] += A[i][k] * B[k][j]

print("\nMatrix Multiplication")

for row in result:
    print(row)
```

#Identity Matrix

```python
n = int(input("Enter size of matrix: "))

for i in range(n):
    for j in range(n):
        if i == j:
            print(1, end=" ")
        else:
            print(0, end=" ")
    print()
```

#Transpose Matrix

```python
rows = int(input("Enter rows: "))
cols = int(input("Enter columns: "))

matrix = []

print("Enter Matrix")

for i in range(rows):
    row = []
    for j in range(cols):
        row.append(int(input()))
    matrix.append(row)

print("\nTranspose Matrix")

for i in range(cols):
    for j in range(rows):
        print(matrix[j][i], end=" ")
    print()
```

#Chessboard Pattern

```python
size = 8

for i in range(size):
    for j in range(size):
        if (i + j) % 2 == 0:
            print("■", end=" ")
        else:
            print("□", end=" ")
    print()
```
