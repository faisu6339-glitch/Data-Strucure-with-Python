## Insertion Sort Explained

**Insertion Sort** is a simple sorting algorithm that builds the final sorted array (or list) one item at a time. It is much less efficient on large lists than more advanced algorithms like quicksort, heapsort, or merge sort. However, it can be quite efficient for small lists or for lists that are already substantially sorted.

### How it Works (Analogy)

Think of how you might sort a hand of playing cards. You hold the cards in your hand, and they are already sorted. When you pick up a new card, you find the correct position for it among the cards you already hold and insert it there. This is essentially what insertion sort does.

### Algorithm Steps

1.  **Start with the second element**: The first element is considered sorted (a list with one element is always sorted).
2.  **Iterate through the unsorted portion**: Take the next element from the unsorted part of the array. Let's call this `key`.
3.  **Compare and shift**: Compare the `key` with elements in the sorted part, moving from right to left.
    *   If an element in the sorted part is greater than the `key`, shift it one position to the right.
    *   Continue this process until you find an element that is less than or equal to the `key`, or you reach the beginning of the array.
4.  **Insert the key**: Place the `key` in the position where the shifting stopped.
5.  **Repeat**: Continue steps 2-4 until all elements have been moved from the unsorted part to the sorted part.

### Detailed Algorithm

Let `arr` be an array of `n` elements.

```python
function insertionSort(arr):
    n = length(arr)
    for i from 1 to n-1:  # Iterate from the second element to the last
        key = arr[i]       # The element to be inserted into the sorted part
        j = i - 1          # Start comparing with the element just before 'key'

        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j = j - 1
        arr[j + 1] = key   # Place the key in its correct position
```

### Example

Let's trace `arr = [12, 11, 13, 5, 6]`

**First Pass (i = 1, key = 11)**:
*   `arr` is `[12, 11, 13, 5, 6]`
*   `key = 11`, `j = 0`
*   `11 < 12` is true. Shift `12` to the right: `[12, 12, 13, 5, 6]`
*   `j` becomes `-1`. Loop terminates.
*   Insert `key` (`11`) at `j + 1 = 0`: `[11, 12, 13, 5, 6]`

**Second Pass (i = 2, key = 13)**:
*   `arr` is `[11, 12, 13, 5, 6]`
*   `key = 13`, `j = 1`
*   `13 < 12` is false. Loop terminates.
*   Insert `key` (`13`) at `j + 1 = 2`: `[11, 12, 13, 5, 6]` (no change in this step)

**Third Pass (i = 3, key = 5)**:
*   `arr` is `[11, 12, 13, 5, 6]`
*   `key = 5`, `j = 2`
*   `5 < 13` is true. Shift `13`: `[11, 12, 13, 13, 6]`
*   `j` becomes `1`
*   `5 < 12` is true. Shift `12`: `[11, 12, 12, 13, 6]`
*   `j` becomes `0`
*   `5 < 11` is true. Shift `11`: `[11, 11, 12, 13, 6]`
*   `j` becomes `-1`. Loop terminates.
*   Insert `key` (`5`) at `j + 1 = 0`: `[5, 11, 12, 13, 6]`

**Fourth Pass (i = 4, key = 6)**:
*   `arr` is `[5, 11, 12, 13, 6]`
*   `key = 6`, `j = 3`
*   `6 < 13` is true. Shift `13`: `[5, 11, 12, 13, 13]`
*   `j` becomes `2`
*   `6 < 12` is true. Shift `12`: `[5, 11, 12, 12, 13]`
*   `j` becomes `1`
*   `6 < 11` is true. Shift `11`: `[5, 11, 11, 12, 13]`
*   `j` becomes `0`
*   `6 < 5` is false. Loop terminates.
*   Insert `key` (`6`) at `j + 1 = 1`: `[5, 6, 11, 12, 13]`

The array is now sorted.

### Complexity

*   **Time Complexity:**
    *   **Worst Case:** O(n^2) - Occurs when the array is sorted in reverse order.
    *   **Best Case:** O(n) - Occurs when the array is already sorted. Each element only needs one comparison.
    *   **Average Case:** O(n^2)
*   **Space Complexity:** O(1) - It sorts the array in-place, requiring minimal additional space.

#Program 1: Basic Insertion Sort

````python
 def insertion_sort(arr):
    n = len(arr)

    # Traverse through 1 to len(arr)
    for i in range(1, n):
        key = arr[i] # Current element to be inserted

        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key # Place the key in its correct position

    return arr

# Example usage:
unsorted_list = [12, 11, 13, 5, 6]
sorted_list = insertion_sort(unsorted_list.copy()) # Use .copy() to avoid modifying original list

print(f"Original list: {unsorted_list}")
print(f"Sorted list: {sorted_list}")

unsorted_list_2 = [5, 1, 4, 2, 8]
sorted_list_2 = insertion_sort(unsorted_list_2.copy())
print(f"\nOriginal list: {unsorted_list_2}")
print(f"Sorted list: {sorted_list_2}")
````

#Program 2: Student Marks

````python
def students(marks):

    for i in range(1, len(marks)):

        key = marks[i]
        j = i - 1

        while j >= 0 and marks[j] > key:
            marks[j + 1] = marks[j]
            j -= 1

        marks[j + 1] = key

    return marks


marks = [95, 85, 90, 75, 65]

print(students(marks))
````

#Program 4: Count Shifts

````python
arr = [8, 5, 3, 9, 1]

shifts = 0

for i in range(1, len(arr)):

    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        shifts += 1
        j -= 1

    arr[j + 1] = key

print(arr)
print("Total Shifts:", shifts)
````
