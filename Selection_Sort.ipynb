## Selection Sort Algorithm

Selection Sort is a simple comparison-based sorting algorithm. It works by repeatedly finding the minimum (or maximum) element from the unsorted part of the list and putting it at the beginning (or end) of the sorted part of the list.

### How it works (Algorithm Steps):

1.  **Initialization**: Start with the first element as the `current_index` and consider the entire array as the unsorted portion.

2.  **Find Minimum**: Iterate through the unsorted portion of the array (from `current_index` to the end) to find the element with the smallest value. Keep track of the `min_index` (index of the minimum element found).

3.  **Swap**: If the `min_index` is different from the `current_index`, swap the element at `current_index` with the element at `min_index`. This places the smallest element found in its correct sorted position.

4.  **Advance**: Increment `current_index` by one. The element at the previous `current_index` is now considered part of the sorted portion of the array.

5.  **Repeat**: Repeat steps 2-4 until the entire array is sorted (i.e., `current_index` reaches the second to last element, as the last element will automatically be in its correct place).

### Example:

Let's sort the array `[64, 25, 12, 22, 11]` using Selection Sort:

**Pass 1:**
-   `current_index = 0` (value is 64)
-   Find minimum in `[64, 25, 12, 22, 11]`. The minimum is 11 at index 4.
-   Swap 64 and 11. Array becomes `[11, 25, 12, 22, 64]`.

**Pass 2:**
-   `current_index = 1` (value is 25)
-   Find minimum in `[25, 12, 22, 64]` (unsorted portion). The minimum is 12 at index 2 (relative to original array).
-   Swap 25 and 12. Array becomes `[11, 12, 25, 22, 64]`.

**Pass 3:**
-   `current_index = 2` (value is 25)
-   Find minimum in `[25, 22, 64]` (unsorted portion). The minimum is 22 at index 3.
-   Swap 25 and 22. Array becomes `[11, 12, 22, 25, 64]`.

**Pass 4:**
-   `current_index = 3` (value is 25)
-   Find minimum in `[25, 64]` (unsorted portion). The minimum is 25 at index 3.
-   No swap needed (element is already in place).
-   Array remains `[11, 12, 22, 25, 64]`.

The array is now sorted.

### Characteristics:

*   **Time Complexity**: O(n^2) in all cases (best, average, worst) because it always makes n-1 passes and each pass involves finding the minimum element in the remaining unsorted subarray.
*   **Space Complexity**: O(1) because it only uses a few extra variables for comparison and swapping.
*   **Stability**: Not stable by default (the relative order of equal elements might change).
*   **In-place**: Yes, it sorts the array by modifying the original array directly without needing a temporary array.

Selection sort is generally not efficient for large datasets due to its quadratic time complexity. However, it can be useful for smaller lists or when memory write operations are costly, as it performs a minimum number of swaps.

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        # Assume the current element is the minimum
        min_idx = i

        # Find the minimum element in the remaining unsorted array
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j

        # Swap the found minimum element with the first element of the unsorted part
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Demo
my_list = [64, 25, 12, 22, 11]
print(f"Original list: {my_list}")
sorted_list = selection_sort(my_list)
print(f"Sorted list: {sorted_list}")

my_list_2 = [5, 4, 3, 2, 1]
print(f"Original list 2: {my_list_2}")
sorted_list_2 = selection_sort(my_list_2)
print(f"Sorted list 2: {sorted_list_2}")
```

#Program 2: Selection Sort (Descending)

```python
def selection_sort_desc(arr):
    n = len(arr)

    for i in range(n):
        max_index = i

        for j in range(i + 1, n):
            if arr[j] > arr[max_index]:
                max_index = j

        arr[i], arr[max_index] = arr[max_index], arr[i]

    return arr


numbers = [15, 8, 2, 45, 10]

selection_sort_desc(numbers)

print(numbers)
```

#Program 3: User Input

```python
arr = list(map(int, input("Enter numbers: ").split()))

for i in range(len(arr)):
    min_index = i

    for j in range(i + 1, len(arr)):
        if arr[j] < arr[min_index]:
            min_index = j

    arr[i], arr[min_index] = arr[min_index], arr[i]

print("Sorted Array:", arr)
```

#Program 4: Sort Student Marks

```python
def students(marks):
    n = len(marks)

    for i in range(n - 1):
        min_index = i

        for j in range(i + 1, n):
            if marks[j] < marks[min_index]:
                min_index = j

        if min_index != i:
            marks[i], marks[min_index] = marks[min_index], marks[i]

    return marks


marks = [95, 85, 90, 75, 65]

print("Original Marks:", marks)
sorted_marks = students(marks)
print("Sorted Marks:", sorted_marks)
```

#Program 5: Sort Dictionary by Value

```python
students = {
    "Aman": 88,
    "Rahul": 75,
    "Priya": 95,
    "Riya": 81
}

items = list(students.items())

for i in range(len(items)):
    min_index = i

    for j in range(i + 1, len(items)):
        if items[j][1] < items[min_index][1]:
            min_index = j

    items[i], items[min_index] = items[min_index], items[i]

print(dict(items))
```
