## Quicksort Algorithm Explained

Quicksort is a highly efficient, comparison-based sorting algorithm. It was developed by Tony Hoare in 1959 and is often used for internal sorting (sorting data in primary storage).

### How it Works (Divide and Conquer)

Quicksort follows the **divide and conquer** paradigm. Here's the general idea:

1.  **Divide**: Pick an element from the array, called a **pivot**. Rearrange the array elements in such a way that all elements smaller than the pivot come before it, and all elements greater than the pivot come after it. Elements equal to the pivot can go on either side. After this partitioning, the pivot is in its final sorted position.
2.  **Conquer**: Recursively apply the above steps to the sub-array of elements with smaller values and separately to the sub-array of elements with greater values.
3.  **Combine**: No combination step is needed, as the sub-arrays are sorted in place.

### Detailed Steps:

1.  **Choose a Pivot**: Select an element from the array to be the pivot. The choice of pivot significantly affects the algorithm's performance. Common strategies include:
    *   Always picking the first element.
    *   Always picking the last element.
    *   Picking a random element.
    *   Picking the median of three (first, middle, last) elements.

2.  **Partitioning**: Rearrange the array such that:
    *   All elements less than the pivot are placed to its left.
    *   All elements greater than the pivot are placed to its right.
    *   The pivot itself is placed in its correct sorted position.
    This operation is called `partition()` and it returns the index of the pivot after rearrangement.

3.  **Recursion**: Recursively apply Quicksort to the sub-array to the left of the pivot and to the sub-array to the right of the pivot.

### Example of Partitioning (Lomuto Partition Scheme)

Let's consider an array `arr = [10, 80, 30, 90, 40, 50, 70]` and choose the last element `70` as the pivot.

1.  `pivot = 70`.
2.  Initialize `i` (index of smaller element) to `low - 1` (let's say `low = 0`, so `i = -1`).
3.  Iterate `j` from `low` to `high - 1`:
    *   If `arr[j] <= pivot`:
        *   Increment `i`.
        *   Swap `arr[i]` and `arr[j]`.

**Trace:**

Initial `arr = [10, 80, 30, 90, 40, 50, 70]`, `pivot = 70`, `i = -1`.

*   `j = 0`, `arr[0] = 10` (10 <= 70). `i` becomes `0`. Swap `arr[0]` and `arr[0]` (no change). `arr = [10, 80, 30, 90, 40, 50, 70]`.
*   `j = 1`, `arr[1] = 80` (80 > 70). No swap.
*   `j = 2`, `arr[2] = 30` (30 <= 70). `i` becomes `1`. Swap `arr[1]` and `arr[2]`. `arr = [10, 30, 80, 90, 40, 50, 70]`.
*   `j = 3`, `arr[3] = 90` (90 > 70). No swap.
*   `j = 4`, `arr[4] = 40` (40 <= 70). `i` becomes `2`. Swap `arr[2]` and `arr[4]`. `arr = [10, 30, 40, 90, 80, 50, 70]`.
*   `j = 5`, `arr[5] = 50` (50 <= 70). `i` becomes `3`. Swap `arr[3]` and `arr[5]`. `arr = [10, 30, 40, 50, 80, 90, 70]`.

After loop, `i = 3`. Swap `arr[i+1]` (`arr[4] = 80`) with `arr[high]` (`arr[6] = 70`, the pivot).

Final `arr = [10, 30, 40, 50, 70, 90, 80]`.

The pivot `70` is now at index `4` (which is `i+1`), with all smaller elements to its left and larger elements to its right.

### Python Implementation

```python
def quicksort(arr, low, high):
    if low < high:
        # pi is partitioning index, arr[pi] is now at right place
        pi = partition(arr, low, high)

        # Separately sort elements before partition and after partition
        quicksort(arr, low, pi - 1)
        quicksort(arr, pi + 1, high)

def partition(arr, low, high):
    # Choose the last element as the pivot
    pivot = arr[high]

    # Index of smaller element
    i = low - 1

    for j in range(low, high):
        # If current element is smaller than or equal to pivot
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i] # Swap

    arr[i + 1], arr[high] = arr[high], arr[i + 1] # Swap pivot into its correct position
    return i + 1

# Example Usage:
arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
quicksort(arr, 0, n - 1)
print("Sorted array is:", arr)

arr2 = [3, 0, 2, 5, -1, 4]
n2 = len(arr2)
quicksort(arr2, 0, n2 - 1)
print("Sorted array is:", arr2)
```

### Time and Space Complexity

*   **Time Complexity:**
    *   **Best Case: O(n log n)** - Occurs when the pivot always divides the array into two (nearly) equal halves. This is similar to Merge Sort.
    *   **Average Case: O(n log n)** - When the pivot choice is reasonably good on average.
    *   **Worst Case: O(n^2)** - Occurs when the pivot always results in highly unbalanced partitions (e.g., the array is already sorted or reverse sorted, and the first/last element is always chosen as the pivot). This can be mitigated by choosing a good pivot (e.g., random pivot or median-of-three).

*   **Space Complexity:**
    *   **O(log n)** (average case) - Due to the recursive call stack. The depth of the recursion is `log n` in the best and average cases.
    *   **O(n)** (worst case) - In the worst-case scenario (unbalanced partitions), the recursion depth can go up to `n`.

### Advantages of Quicksort

*   **Fast**: On average, it is faster than other `O(n log n)` algorithms like Merge Sort.
*   **In-place sorting**: It sorts the array without requiring significant extra space (unlike Merge Sort, which typically requires `O(n)` extra space).
*   **Cache-friendly**: Its memory access patterns can be more cache-friendly than other sorting algorithms, leading to better performance in practice.

### Disadvantages of Quicksort

*   **Worst-case performance**: `O(n^2)` can be a concern if pivots are consistently chosen poorly, though this is rare with good pivot selection strategies.
*   **Not stable**: The relative order of equal elements is not preserved.
*   **Recursive**: Can lead to stack overflow for very large arrays in languages without tail call optimization or with deep recursion limits, although iterative versions exist.

```python
def partition(arr, low, high):
    pivot = arr[high]          # Last element as pivot
    i = low - 1                # Index of smaller element

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1


def quick_sort(arr, low, high):
    if low < high:
        pivot_index = partition(arr, low, high)

        quick_sort(arr, low, pivot_index - 1)
        quick_sort(arr, pivot_index + 1, high)


numbers = [10, 7, 8, 9, 1, 5]

quick_sort(numbers, 0, len(numbers)-1)

print(numbers)
```

```python
def partition(arr, low, high):
    pivot = arr[low]

    i = low + 1
    j = high

    while True:

        while i <= high and arr[i] <= pivot:
            i += 1

        while arr[j] > pivot:
            j -= 1

        if i < j:
            arr[i], arr[j] = arr[j], arr[i]
        else:
            break

    arr[low], arr[j] = arr[j], arr[low]

    return j


def quick_sort(arr, low, high):
    if low < high:
        p = partition(arr, low, high)

        quick_sort(arr, low, p - 1)
        quick_sort(arr, p + 1, high)


arr = [50, 70, 40, 20, 60, 10]

quick_sort(arr, 0, len(arr) - 1)

print(arr)
```

```python
def quick_sort(arr):

    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]

    left = [x for x in arr if x < pivot]

    middle = [x for x in arr if x == pivot]

    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)


numbers = [10, 7, 8, 9, 1, 5]

print(quick_sort(numbers))
```
