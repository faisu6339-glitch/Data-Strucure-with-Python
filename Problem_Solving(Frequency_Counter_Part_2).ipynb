#1. Maximum Sum Subarray of Size K
Problem Statement

Given an array of integers and an integer k, find the maximum sum of any contiguous subarray of size k.

Example

````python
Input:
arr = [2,1,5,1,3,2]
k = 3

Output:
9

Explanation:
Subarrays:
[2,1,5] = 8
[1,5,1] = 7
[5,1,3] = 9
[1,3,2] = 6
Maximum = 9
````

````python
def max_sum_subarray(arr, k):

    window_sum = sum(arr[:k])
    max_sum = window_sum

    for i in range(k, len(arr)):
        window_sum += arr[i]
        window_sum -= arr[i-k]

        max_sum = max(max_sum, window_sum)

    return max_sum


# Test Cases
print(max_sum_subarray([2,1,5,1,3,2],3))      # 9
print(max_sum_subarray([1,2,3,4,5],2))        # 9
print(max_sum_subarray([5,5,5],2))            # 10
````

### Explanation of `max_sum_subarray` Function

This function uses the **sliding window technique** to efficiently find the maximum sum of a contiguous subarray of a given size `k`.

#### Step-by-Step Dry Run Example: `max_sum_subarray([2,1,5,1,3,2], 3)`

Let's trace the execution with `arr = [2,1,5,1,3,2]` and `k = 3`.

##### Step 1: Initialize `window_sum`

```python
window_sum = sum(arr[:k])
```

Since `k = 3`, `arr[:k]` is `arr[:3]`, which evaluates to `[2,1,5]`.

So, `window_sum = 2 + 1 + 5 = 8`.

Current Window: `[2,1,5]`
`window_sum = 8`

##### Step 2: Initialize `max_sum`

```python
max_sum = window_sum
```

Initially, `max_sum = 8`.

Current situation:
Window: `[2,1,5]`
`window_sum = 8`
`max_sum = 8`

##### Step 3: Loop through the array (`for i in range(k, len(arr))`)

Since `k = 3` and `len(arr) = 6`, the loop will run for `i` in `range(3, 6)`, meaning `i = 3, 4, 5`.

###### First Iteration: `i = 3`

*   **Current window:** `[2,1,5]` (sum = 8)
*   **We want the next window:** `[1,5,1]`

Instead of recalculating `1+5+1`, the sliding window approach updates the `window_sum`:

1.  **Add new element:** `window_sum += arr[i]`
    *   `arr[3]` is `1`.
    *   `window_sum` becomes `8 + 1 = 9`.

2.  **Remove old element:** `window_sum -= arr[i-k]`
    *   `i - k` is `3 - 3 = 0`.
    *   `arr[0]` is `2`.
    *   `window_sum` becomes `9 - 2 = 7`.

    Now, the current window is `[1,5,1]` with `sum = 7`.

3.  **Update `max_sum`:** `max_sum = max(max_sum, window_sum)`
    *   `max(8, 7)` is `8`.
    *   `max_sum` remains `8`.

###### Second Iteration: `i = 4`

*   **Current window:** `[1,5,1]` (sum = 7)
*   **We want the next window:** `[5,1,3]`

1.  **Add new element:** `arr[4]` is `3`.
    *   `window_sum` becomes `7 + 3 = 10`.

2.  **Remove old element:** `i - k` is `4 - 3 = 1`.
    *   `arr[1]` is `1`.
    *   `window_sum` becomes `10 - 1 = 9`.

    Now, the current window is `[5,1,3]` with `sum = 9`.

3.  **Update `max_sum`:** `max(8, 9)` is `9`.
    *   `max_sum` becomes `9`.

###### Third Iteration: `i = 5`

*   **Current window:** `[5,1,3]` (sum = 9)
*   **We want the next window:** `[1,3,2]`

1.  **Add new element:** `arr[5]` is `2`.
    *   `window_sum` becomes `9 + 2 = 11`.

2.  **Remove old element:** `i - k` is `5 - 3 = 2`.
    *   `arr[2]` is `5`.
    *   `window_sum` becomes `11 - 5 = 6`.

    Now, the current window is `[1,3,2]` with `sum = 6`.

3.  **Update `max_sum`:** `max(9, 6)` is `9`.
    *   `max_sum` remains `9`.

The loop ends here.

##### Final Result

The function returns `max_sum`, which is `9`.

##### Complete Dry Run Table

| Iteration | `i` | Window  | Added | Removed | `window_sum` | `max_sum` |
|-----------|-----|---------|-------|---------|--------------|-----------|
| Initial   | -   | `[2,1,5]` | -     | -       | 8            | 8         |
| 1         | 3   | `[1,5,1]` | 1     | 2       | 7            | 8         |
| 2         | 4   | `[5,1,3]` | 3     | 1       | 9            | 9         |
| 3         | 5   | `[1,3,2]` | 2     | 5       | 6            | 9         |

##### Why `arr[i-k]`?

This `arr[i-k]` term is crucial for the sliding window. It always refers to the element that is exiting the window as the window slides one position to the right.

Let's visualize with `k=3`:

*   **Initial Window:** `[**2**, 1, 5]`
    *   When `i=3`, `i-k = 0`. `arr[0]` is `2`, which is the element leaving.
    
*   **Next Window:** `[1, **5**, 1]` (from `[1, 5, 1]`) (leaving `2`, adding `1`)
    *   When `i=4`, `i-k = 1`. `arr[1]` is `1`, which is the element leaving.

*   **Next Window:** `[5, **1**, 3]` (from `[5, 1, 3]`) (leaving `1`, adding `3`)
    *   When `i=5`, `i-k = 2`. `arr[2]` is `5`, which is the element leaving.

This pattern confirms that `arr[i-k]` consistently targets the element that needs to be removed from the `window_sum` to maintain the correct sum for the current sliding window.

# 2. Minimum Sum Subarray of Size K
Problem Statement

Find the minimum sum among all subarrays of size k.

````python
Input:
arr=[2,1,5,1,3,2]

k=3

Output:
6
````

````python
def min_sub_array(arr,k):
  window_sum=sum(arr[:k])
  min_sum=window_sum

  for i in range(k,len(arr)):
    window_sum +=arr[i]
    window_sum -=arr[i-k]

    min_sum=min(min_sum,window_sum)

  return min_sum

print(min_sub_array([2,1,5,1,3,2],3))
````

# 3. Average of Every Subarray of Size K
Problem Statement

Return the average of every window of size k.

Example

````python
Input:
[1,3,2,6,-1,4]

k=3

Output:
[2.0,3.67,2.33,3.0]
````

````python
def find_averages(arr,size):
  result=[]
  window_sum=sum(arr[:size])
  result.append(window_sum/size)

  for i in range(size,len(arr)):
    window_sum=window_sum-arr[i - size]+arr[i]
    result.append(window_sum/size)

  return result

print(find_averages([1,3,2,6,-1,4],3))
````

# 4. Maximum Average Subarray
Problem Statement

Find the maximum average of all subarrays having size k.

Example

````python
Input:
[1,12,-5,-6,50,3]

k=4

Output:
12.75
````

````python
def maximum_average(arr,k):
  window_sum=sum(arr[:k])
  max_average=window_sum/k

  for i in range(k,len(arr)):

    window_sum+=arr[i]
    window_sum-=arr[i-k]

    max_average=max(max_average,window_sum/k)

  return max_average

print(maximum_average([1,12,-5,-6,50,3],4))
````

````python
def max_vowel(s, t):
  vowels = {"a", "e", "i", "o", "u"}
  current_vowel_count = 0

  # Calculate vowel count for the first window of size t
  for i in range(t):
    if s[i] in vowels:
      current_vowel_count += 1

  max_vowel_count = current_vowel_count

  # Slide the window across the rest of the string
  for i in range(t, len(s)):
    # Add the new character entering the window
    if s[i] in vowels:
      current_vowel_count += 1
    # Remove the character leaving the window
    if s[i - t] in vowels:
      current_vowel_count -= 1

    # Update max_vowel_count if the current window has more vowels
    max_vowel_count = max(max_vowel_count, current_vowel_count)

  return max_vowel_count

# Test Cases
print(f"'abciiidef', k=3 -> {max_vowel('abciiidef', 3)}") # Expected: 3 ('iii')
print(f"'aeiou', k=2 -> {max_vowel('aeiou', 2)}")       # Expected: 2 ('ae', 'ei', 'io', 'ou')
print(f"'leetcode', k=3 -> {max_vowel('leetcode', 3)}")    # Expected: 2 ('eet', 'ode')
print(f"'rhythms', k=4 -> {max_vowel('rhythms', 4)}")    # Expected: 0
print(f"'abc', k=1 -> {max_vowel('abc', 1)}")         # Expected: 1
print(f"'weallloveyou', k=7 -> {max_vowel('weallloveyou', 7)}") # Expected: 4
````

### Explanation of `max_vowel` Function

The `max_vowel` function is designed to find the maximum number of vowels in any contiguous substring (window) of a given length `t` within a string `s`. It efficiently achieves this using the **sliding window technique**.

#### How it works:

1.  **Define Vowels:** It first creates a set of vowels for quick lookup.
2.  **Initial Window Calculation:** It calculates the number of vowels in the *first* window of size `t`. This `current_vowel_count` is also set as the initial `max_vowel_count`.
3.  **Slide the Window:** It then iterates through the rest of the string, effectively sliding the window one character at a time.
    *   For each step, it **adds** the count of the new character entering the window (if it's a vowel).
    *   It **subtracts** the count of the character leaving the window (if it was a vowel).
    *   This way, `current_vowel_count` always reflects the number of vowels in the current window without recalculating from scratch.
4.  **Update Maximum:** After each slide, it compares the `current_vowel_count` with `max_vowel_count` and updates `max_vowel_count` if the current window has more vowels.
5.  **Return Result:** Finally, it returns the `max_vowel_count` found.

#### Logic behind `s[i-t]`

In the sliding window approach, `s[i-t]` represents the character that is **leaving** the window as it slides one position to the right. The window's size `t` remains constant.

Let's break it down:

*   When the loop variable `i` is at a certain index, `s[i]` is the character **entering** the window from the right.
*   To keep the window size `t` constant, one character must **leave** the window from the left.
*   The index of this leaving character is precisely `i - t`.

Consider our example `s = 'abciiidef'` and `t = 3`:

1.  **Initial Window:** `a b c`
    *   `i` starts at `3` (the first character *outside* the initial window to be included).
    *   `s[3]` is `'i'`, which enters the window.
    *   The character that must leave is at `s[3-3] = s[0]`, which is `'a'`. This is the element at the beginning of the previous window.

2.  **Next Window:** `b c i`
    *   `i` is now `4`.
    *   `s[4]` is `'i'`, which enters the window.
    *   The character that must leave is at `s[4-3] = s[1]`, which is `'b'`. This was the first character of the `bci` window.

By checking if `s[i-t]` is a vowel, we can efficiently decrement `current_vowel_count` if the character leaving the window was indeed a vowel. This avoids re-counting vowels for the entire window in every iteration, making the algorithm much faster, especially for large strings.
