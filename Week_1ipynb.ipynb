#Linear Search

**Program 1**

```python
def student_marks(marks, target):
    for i in range(len(marks)):
        if marks[i] == target:
            print("Yes, student with", target, "marks found.")
            return i   # Return the index where the marks are found

    print("No student with", target, "marks found.")
    return -1
```

```python
marks = [95, 85, 90, 75, 65]

target = 90

result = student_marks(marks, target)

if result != -1:
    print("Found at index:", result)
```

#Binary Search

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
```

```python
arr = [10, 20, 30, 40, 50, 60, 70]
target = 40

result = binary_search(arr, target)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

**Program 2**

```python
def student_roll(roll, target):
    low = 0
    high = len(roll) - 1

    while low <= high:
        mid = (low + high) // 2

        if roll[mid] == target:
            return mid
        elif roll[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
```

```python
roll = [101, 102, 103, 104, 105, 106]

target = 104

result = student_roll(roll, target)

if result != -1:
    print("Student found at index:", result)
else:
    print("Student not found")
```

**Program 3**

```python
def first_occurrence(arr, target):
    left = 0
    right = len(arr) - 1
    result = -1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            result = mid          # Store the index
            right = mid - 1       # Continue searching on the left
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return result


arr = [10, 20, 20, 20, 30, 40]
target = 20

print(first_occurrence(arr, target))
```

**Program 4**

```python
def last_occurrence(arr, target):
    left = 0
    right = len(arr) - 1
    result = -1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            result = mid          # Store the index
            left = mid + 1        # Continue searching on the right
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return result


arr = [10, 20, 20, 20, 30, 40]
target = 20

print(last_occurrence(arr, target))
```

**Program 5**

```python
def count_occurrence(arr, target):
    first = first_occurrence(arr, target)

    if first == -1:
        return 0

    last = last_occurrence(arr, target)

    return last - first + 1


arr = [10, 20, 20, 20, 30, 40]
target = 20

print("Count =", count_occurrence(arr, target))
```

**Program 6**

```python
def find_peak(arr):
    left = 0
    right = len(arr) - 1

    while left < right:
        mid = (left + right) // 2

        if arr[mid] < arr[mid + 1]:
            left = mid + 1
        else:
            right = mid

    return left


arr = [10, 20, 50, 40, 30]

index = find_peak(arr)

print("Peak Index:", index)
print("Peak Element:", arr[index])
```
