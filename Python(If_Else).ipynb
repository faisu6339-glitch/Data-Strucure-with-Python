#7.Gross Salary

```python
# ==========================================
#          GROSS SALARY CALCULATOR
# ==========================================

print("=" * 60)
print("           GROSS SALARY CALCULATOR")
print("=" * 60)

# Employee Details
name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
department = input("Enter Department: ")
designation = input("Enter Designation: ")

# Basic Salary
basic_salary = float(input("Enter Basic Salary (₹): "))

# Calculate HRA and DA
if basic_salary <= 20000:
    hra = basic_salary * 0.20      # 20%
    da = basic_salary * 0.50       # 50%

elif basic_salary <= 50000:
    hra = basic_salary * 0.30      # 30%
    da = basic_salary * 0.60       # 60%

else:
    hra = basic_salary * 0.40      # 40%
    da = basic_salary * 0.70       # 70%

# Medical Allowance
medical = 3000

# Travel Allowance
travel = 2000

# Bonus
if basic_salary >= 50000:
    bonus = 10000
elif basic_salary >= 30000:
    bonus = 5000
else:
    bonus = 2000

# Gross Salary
gross_salary = basic_salary + hra + da + medical + travel + bonus

# Tax Calculation
if gross_salary >= 100000:
    tax = gross_salary * 0.20

elif gross_salary >= 50000:
    tax = gross_salary * 0.10

else:
    tax = gross_salary * 0.05

# Net Salary
net_salary = gross_salary - tax

# Display Salary Slip
print("\n" + "=" * 60)
print("                 SALARY SLIP")
print("=" * 60)

print("Employee Name      :", name)
print("Employee ID        :", emp_id)
print("Department         :", department)
print("Designation        :", designation)

print("-" * 60)
print("Basic Salary       : ₹", basic_salary)
print("HRA                : ₹", hra)
print("DA                 : ₹", da)
print("Medical Allowance  : ₹", medical)
print("Travel Allowance   : ₹", travel)
print("Bonus              : ₹", bonus)
print("-" * 60)

print("Gross Salary       : ₹", gross_salary)
print("Tax Deduction      : ₹", tax)
print("Net Salary         : ₹", net_salary)

print("=" * 60)

# Salary Grade
if net_salary >= 100000:
    print("Grade : A+ (Excellent Salary)")

elif net_salary >= 70000:
    print("Grade : A")

elif net_salary >= 50000:
    print("Grade : B")

elif net_salary >= 30000:
    print("Grade : C")

else:
    print("Grade : D")

print("=" * 60)
print("Thank You!")
```

#8.Electricity Bill

```python
# ==========================================
#       ELECTRICITY BILL CALCULATOR
# ==========================================

print("=" * 60)
print("          ELECTRICITY BILL SYSTEM")
print("=" * 60)

# Customer Details
customer_name = input("Enter Customer Name: ")
customer_id = input("Enter Customer ID: ")
city = input("Enter City: ")

# Input
previous_reading = int(input("Enter Previous Meter Reading: "))
current_reading = int(input("Enter Current Meter Reading: "))

# Calculate Units
units = current_reading - previous_reading

# Validate Reading
if units < 0:
    print("\n❌ Invalid Meter Reading!")

else:

    # Bill Calculation
    if units <= 100:
        bill = units * 5

    elif units <= 200:
        bill = (100 * 5) + ((units - 100) * 7)

    elif units <= 300:
        bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

    else:
        bill = (100 * 5) + (100 * 7) + (100 * 10) + ((units - 300) * 12)

    # Fixed Charges
    fixed_charge = 150

    # Electricity Duty (5%)
    duty = bill * 0.05

    # Total Bill
    total_bill = bill + fixed_charge + duty

    # Late Payment Surcharge
    late_payment = input("Late Payment? (yes/no): ").lower()

    if late_payment == "yes":
        surcharge = total_bill * 0.10
    else:
        surcharge = 0

    final_bill = total_bill + surcharge

    # Display Bill
    print("\n" + "=" * 60)
    print("               ELECTRICITY BILL")
    print("=" * 60)

    print("Customer Name      :", customer_name)
    print("Customer ID        :", customer_id)
    print("City               :", city)
    print("Units Consumed     :", units)

    print("-" * 60)
    print("Energy Charge      : ₹", bill)
    print("Fixed Charge       : ₹", fixed_charge)
    print("Electricity Duty   : ₹", round(duty, 2))
    print("Late Fee           : ₹", round(surcharge, 2))
    print("-" * 60)

    print("Final Bill Amount  : ₹", round(final_bill, 2))

    # Bill Category
    print("\nBill Category")

    if final_bill >= 10000:
        print("⭐⭐⭐⭐ Very High Consumption")

    elif final_bill >= 5000:
        print("⭐⭐⭐ High Consumption")

    elif final_bill >= 2000:
        print("⭐⭐ Moderate Consumption")

    else:
        print("⭐ Low Consumption")

    print("=" * 60)
    print("Thank You! Visit Again.")
```

#8.Calculator using If-Else

```python
# ==========================================
#          SIMPLE CALCULATOR
# ==========================================

print("=" * 50)
print("         SIMPLE CALCULATOR")
print("=" * 50)

# Input
num1 = float(input("Enter First Number: "))
num2 = float(input("Enter Second Number: "))

print("\nChoose an Operation")
print("1. Addition (+)")
print("2. Subtraction (-)")
print("3. Multiplication (*)")
print("4. Division (/)")
print("5. Modulus (%)")
print("6. Floor Division (//)")
print("7. Exponentiation (**)")

choice = int(input("\nEnter Your Choice (1-7): "))

print("\n" + "=" * 50)

# Decision Making
if choice == 1:
    result = num1 + num2
    print("Addition =", result)

elif choice == 2:
    result = num1 - num2
    print("Subtraction =", result)

elif choice == 3:
    result = num1 * num2
    print("Multiplication =", result)

elif choice == 4:
    if num2 != 0:
        result = num1 / num2
        print("Division =", result)
    else:
        print("❌ Division by zero is not allowed.")

elif choice == 5:
    if num2 != 0:
        result = num1 % num2
        print("Modulus =", result)
    else:
        print("❌ Modulus by zero is not allowed.")

elif choice == 6:
    if num2 != 0:
        result = num1 // num2
        print("Floor Division =", result)
    else:
        print("❌ Floor division by zero is not allowed.")

elif choice == 7:
    result = num1 ** num2
    print("Exponentiation =", result)

else:
    print("❌ Invalid Choice!")

print("=" * 50)
print("Thank You for Using the Calculator!")
```

#9.ATM Machine

```python
# ==========================================
#           ATM MACHINE SYSTEM
# ==========================================

print("=" * 60)
print("             WELCOME TO XYZ BANK ATM")
print("=" * 60)

# Account Details
account_holder = input("Enter Account Holder Name: ")
pin = int(input("Enter 4-Digit ATM PIN: "))

# Assume Correct PIN
correct_pin = 1234
balance = 50000

# PIN Verification
if pin == correct_pin:

    print("\nLogin Successful")
    print("=" * 60)

    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Change PIN")
    print("5. Exit")

    choice = int(input("\nEnter Your Choice: "))

    # Check Balance
    if choice == 1:
        print("\nCurrent Balance: ₹", balance)

    # Deposit
    elif choice == 2:
        amount = float(input("Enter Deposit Amount: ₹"))

        if amount > 0:
            balance = balance + amount
            print("Deposit Successful")
            print("Updated Balance: ₹", balance)
        else:
            print("Invalid Amount")

    # Withdraw
    elif choice == 3:
        amount = float(input("Enter Withdrawal Amount: ₹"))

        if amount <= 0:
            print("Invalid Amount")

        elif amount > balance:
            print("Insufficient Balance")

        else:
            balance = balance - amount
            print("Please Collect Your Cash")
            print("Remaining Balance: ₹", balance)

    # Change PIN
    elif choice == 4:
        new_pin = int(input("Enter New 4-Digit PIN: "))
        confirm_pin = int(input("Confirm New PIN: "))

        if new_pin == confirm_pin:
            correct_pin = new_pin
            print("PIN Changed Successfully")
        else:
            print("PIN Does Not Match")

    # Exit
    elif choice == 5:
        print("Thank You for Using Our ATM")

    else:
        print("Invalid Choice")

else:
    print("\nIncorrect PIN")
    print("Transaction Cancelled")

print("=" * 60)
print("Visit Again!")
```
