### Variables in Python

In Python, a variable is a named storage location that holds a value. You can think of it as a container that you can put data into. You give the container a name, and then you can refer to that name to access or change the data it holds.

Here's how it works:

*   **Declaration and Assignment**: You create a variable the moment you first assign a value to it using the `=` operator. For example, `num = 10` creates a variable named `num` and assigns it the integer value `10`.
*   **Dynamic Typing**: Python is dynamically typed, which means you don't need to declare the variable's type (like integer, string, etc.) before using it. The type is inferred from the value you assign.
*   **Reassignment**: You can change the value a variable holds at any time, even to a different type. For instance, after `num = 10`, you could later do `num = 'hello'`.

### Different Ways to Assign Variables in Python

```python
# 1. Single Assignment
# This is the most common way to assign a value to a single variable.
x = 10
print(f"Single Assignment: x = {x}")
```

```python
# 2. Multiple Assignments (Tuple Unpacking)
# You can assign multiple values to multiple variables on a single line.
a, b, c = 1, 2, 3
print(f"Multiple Assignments: a = {a}, b = {b}, c = {c}")

# This is also commonly used for swapping variable values
p, q = 100, 200
print(f"Before swap: p = {p}, q = {q}")
p, q = q, p
print(f"After swap: p = {p}, q = {q}")
```

```python
# 3. Chained Assignment
# You can assign the same value to multiple variables simultaneously.
apple = orange = banana = 50
print(f"Chained Assignment: apple = {apple}, orange = {orange}, banana = {banana}")
```

```python
# 4. Augmented Assignment (Compound Assignment Operators)
# These are shorthand operators that combine an operation with an assignment.
# Examples: `+=`, `-=`, `*=`, `/=`, `%=`, `**=` etc.

value = 10
print(f"Initial value: {value}")

value += 5 # Equivalent to value = value + 5
print(f"After += 5: {value}")

value *= 2 # Equivalent to value = value * 2
print(f"After *= 2: {value}")

message = "Hello"
message += " World"
print(f"String concatenation with +=: {message}")
```

#Reverse a Number

```python
num=int(input("Enter the number"))

rev=0
temp=num

while temp>0:
  digits=temp %10
  rev=rev*10+digits
  temp=temp // 10
print("reverse of",num,"is",rev)
```

#Palindrome Number

```python
num=int(input("Enter the numbers"))

original=num
rev=0

while num > 0:
  digits=num%10
  rev=rev*10+digits
  num=num//10

if original==rev:
  print(original,"is a palindrome number")
else:
  print(original,"is not a palindrome number")
```

#Armstrong Number

```python
num = int(input("Enter a number: "))

original = num
power = len(str(num))
total = 0

while num > 0:
    digit = num % 10
    total = total + digit ** power
    num = num // 10

if original == total:
    print(original, "is an Armstrong Number")
else:
    print(original, "is not an Armstrong Number")
```

#Perfect number

```python
num = int(input("Enter a number: "))

total = 0

for i in range(1, num):
    if num % i == 0:
        total += i

if total == num:
    print(num, "is a Perfect Number")
else:
    print(num, "is not a Perfect Number")
```
