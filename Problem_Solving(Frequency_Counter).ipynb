## Problem Solving Pattern: Frequency Counter

### What is the Frequency Counter Pattern?
The Frequency Counter pattern uses objects (or hash maps/dictionaries in Python) to collect the frequencies of values in arrays, strings, or other data structures. This approach can often avoid the need for nested loops or `n^2` comparisons, leading to more efficient `O(n)` solutions for certain types of problems.

### When to use it?
*   When you need to compare two data structures (e.g., two arrays, two strings) to see if they contain the same elements with the same frequencies, or if one is a permutation of the other.
*   When dealing with anagrams.
*   When checking if one array's elements are the squared versions of another array's elements, maintaining order and frequency.

### How it works:
1.  **Create Frequency Maps:** Iterate through the input(s) and build one or more frequency maps (dictionaries in Python). Each map will store elements as keys and their counts as values.
2.  **Compare Maps:** Compare the frequency maps. This comparison can involve:
    *   Checking if corresponding keys exist in both maps.
    *   Checking if the values (counts) for those keys match.

### Example Problem: "Same"

**Problem Statement:** Write a function called `same` which accepts two arrays. The function should return `True` if every value in the first array has its corresponding value squared in the second array. The frequency of values must be the same.

**Example Cases:**
*   `same([1,2,3], [4,1,9])` should return `True`
*   `same([1,2,3], [1,9])` should return `False` (missing `4`)
*   `same([1,2,1], [4,4,1])` should return `False` (frequencies don't match: `1` appears twice in the first, but `1` appears once in the second, and `4` appears twice)

### Naive Solution (Incorrect and Inefficient):

```python
def same_naive(arr1, arr2):
    if len(arr1) != len(arr2):
        return False
    for val in arr1:
        # This will search through arr2 for each element in arr1
        # leading to O(n^2) complexity due to list.index()
        try:
            idx = arr2.index(val ** 2)
            arr2.pop(idx) # Remove the found element to handle duplicates
        except ValueError:
            return False
    return True
```

**Why it's bad:** The `list.index()` and `list.pop()` operations inside a loop make this `O(n^2)` at best, and it doesn't correctly handle frequencies if `pop` is not used carefully.

### Frequency Counter Solution (Efficient):

```python
def same(arr1, arr2):
    if len(arr1) != len(arr2):
        return False

    frequency_counter1 = {}
    frequency_counter2 = {}

    # Populate frequency_counter1
    for val in arr1:
        frequency_counter1[val] = frequency_counter1.get(val, 0) + 1

    # Populate frequency_counter2
    for val in arr2:
        frequency_counter2[val] = frequency_counter2.get(val, 0) + 1

    # Compare the two frequency counters
    for key in frequency_counter1:
        # Check if the squared key exists in the second counter
        if (key ** 2) not in frequency_counter2:
            return False
        # Check if the frequency of the squared key matches
        if frequency_counter2[key ** 2] != frequency_counter1[key]:
            return False

    return True

# Test Cases:
print(f"same([1,2,3], [4,1,9]): {same([1,2,3], [4,1,9])}")        # Expected: True
print(f"same([1,2,3], [1,9]): {same([1,2,3], [1,9])}")          # Expected: False
print(f"same([1,2,1], [4,4,1]): {same([1,2,1], [4,4,1])}")      # Expected: False
print(f"same([2,3,2], [4,9,4]): {same([2,3,2], [4,9,4])}")      # Expected: True
```

### Analysis of Frequency Counter Solution:
*   **Time Complexity:** `O(n)` (where `n` is the length of the arrays). We iterate through `arr1` once, `arr2` once, and then iterate through the keys of `frequency_counter1` (at most `n` unique keys). Dictionary lookups and insertions are, on average, `O(1)`.
*   **Space Complexity:** `O(n)`. We create two dictionaries that, in the worst case (all unique elements), will store `n` key-value pairs each.

### Another Common Example: Anagrams

**Problem Statement:** Given two strings, write a function to determine if the second string is an anagram of the first. An anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

```python
def is_anagram(str1, str2):
    if len(str1) != len(str2):
        return False

    frequency_counter_str1 = {}
    frequency_counter_str2 = {}

    for char in str1:
        frequency_counter_str1[char] = frequency_counter_str1.get(char, 0) + 1

    for char in str2:
        frequency_counter_str2[char] = frequency_counter_str2.get(char, 0) + 1

    for key in frequency_counter_str1:
        if key not in frequency_counter_str2:
            return False
        if frequency_counter_str1[key] != frequency_counter_str2[key]:
            return False

    return True

# Test Cases:
print(f"is_anagram('anagram', 'nagaram'): {is_anagram('anagram', 'nagaram')}") # Expected: True
print(f"is_anagram('rat', 'car'): {is_anagram('rat', 'car')}")           # Expected: False
print(f"is_anagram('awesome', 'awesom'): {is_anagram('awesome', 'awesom')}") # Expected: False
print(f"is_anagram('', ''): {is_anagram('', '')}")               # Expected: True
```

### Key Takeaways for Frequency Counter:
*   **Efficiency:** It's great for optimizing problems that might otherwise involve `O(n^2)` comparisons.
*   **Readability:** The code often becomes clearer by explicitly tracking counts rather than modifying lists in place.
*   **Versatility:** Applicable to various data types as long as you can hash the elements (e.g., numbers, characters, tuples).

```python
def check_doubled_frequency(arr1,arr2):
  if len(arr1)!=len(arr2):
    return False
  target_counts={}

  for x in arr1:
    double_value=x * 2
    target_counts[double_value]=target_counts.get(double_value,0)+1

  for y in arr2:
    if y not in target_counts or target_counts[y]==0:
      return False

    target_counts[y]-=1
  return True

# Test Cases:
print(f"check_doubled_frequency([1,2,3], [2,4,6]): {check_doubled_frequency([1,2,3], [2,4,6])}") # Expected: True
print(f"check_doubled_frequency([1,2,3], [2,6,4]): {check_doubled_frequency([1,2,3], [2,6,4])}") # Expected: True (order doesn't matter)
print(f"check_doubled_frequency([1,2,3], [2,4,5]): {check_doubled_frequency([1,2,3], [2,4,5])}") # Expected: False (5 is not a doubled value)
print(f"check_doubled_frequency([1,2,1], [2,4,2]): {check_doubled_frequency([1,2,1], [2,4,2])}") # Expected: True
print(f"check_doubled_frequency([1,2,1], [2,4,4]): {check_doubled_frequency([1,2,1], [2,4,4])}") # Expected: False (frequencies don't match for 2)
print(f"check_doubled_frequency([1,2,3], [2,4]): {check_doubled_frequency([1,2,3], [2,4])}")   # Expected: False (lengths don't match)
```

### Explanation of Dictionary Logic for Duplicates and Dry Run

The `target_counts` dictionary is crucial for handling duplicate elements. Here's how it works:

1.  **Populating `target_counts` (first loop - `arr1`):**
    *   When we iterate through `arr1`, for each `x`, we calculate its `double_value`.
    *   `target_counts[double_value] = target_counts.get(double_value, 0) + 1` does the following:
        *   If `double_value` is *not yet* a key in `target_counts`, `target_counts.get(double_value, 0)` returns `0`. We then add `1`, so `target_counts[double_value]` becomes `1`.
        *   If `double_value` *is already* a key, `target_counts.get(double_value, 0)` returns its current count. We then add `1` to that count, effectively incrementing the frequency.
    *   This ensures that `target_counts` stores the *exact frequency* of each doubled value that *should* be present in `arr2`.

2.  **Checking `arr2` against `target_counts` (second loop - `arr2`):**
    *   When we iterate through `arr2`, for each `y`, we check if it exists as a key in `target_counts` and if its count is greater than `0` (`target_counts[y] == 0`). If either condition is false, it means `y` was either not the double of any element in `arr1`, or its frequency in `arr2` exceeds what was expected from `arr1`.
    *   If `y` passes the check, we decrement its count: `target_counts[y] -= 1`. This 'consumes' one instance of the expected doubled value.

**Dry Run Example: `check_doubled_frequency([1, 2, 1], [2, 4, 2])`**

*   **Initialization:** `target_counts = {}`

*   **Processing `arr1 = [1, 2, 1]`:**
    *   `x = 1`: `double_value = 2`. `target_counts` becomes `{2: 1}`.
    *   `x = 2`: `double_value = 4`. `target_counts` becomes `{2: 1, 4: 1}`.
    *   `x = 1`: `double_value = 2`. `target_counts.get(2, 0)` returns `1`. `target_counts[2]` becomes `1 + 1 = 2`. `target_counts` becomes `{2: 2, 4: 1}`.

*   **Processing `arr2 = [2, 4, 2]` against `target_counts = {2: 2, 4: 1}`:**
    *   `y = 2`:
        *   `2` is in `target_counts` and `target_counts[2]` (which is `2`) is not `0`. Condition passes.
        *   `target_counts[2]` becomes `2 - 1 = 1`. `target_counts` is now `{2: 1, 4: 1}`.
    *   `y = 4`:
        *   `4` is in `target_counts` and `target_counts[4]` (which is `1`) is not `0`. Condition passes.
        *   `target_counts[4]` becomes `1 - 1 = 0`. `target_counts` is now `{2: 1, 4: 0}`.
    *   `y = 2`:
        *   `2` is in `target_counts` and `target_counts[2]` (which is `1`) is not `0`. Condition passes.
        *   `target_counts[2]` becomes `1 - 1 = 0`. `target_counts` is now `{2: 0, 4: 0}`.

*   **End of loops:** All checks passed. The function returns `True`.

This method ensures that for every `x` in `arr1`, there is a corresponding `2*x` in `arr2`, and that the counts of these doubled values match exactly.

# 2: Square Every Element in an Array

# Problem Statement

Given an array of integers, return a new array where each element is the square of the corresponding element.

Example

Input : [1, 2, 3, 4]

Output: [1, 4, 9, 16]

Approach
Create an empty list.
Traverse the array once.

Square each element using num ** 2.

Store the result in a new list.

Return the new list.

```python
def square_array(arr):
    result = []

    for num in arr:
        square = num ** 2
        result.append(square)

    return result


numbers = [1, 2, 3, 4]

print("Original Array :", numbers)
print("Squared Array  :", square_array(numbers))
```

# Check Squared Frequency

```python
def check_squared_frequency(arr1,arr2):
  if len(arr1)!=len(arr2):
    return False
  target_counts={}

  for x in arr1:
    square= x ** 2
    target_counts[square]=target_counts.get(square,0)+1

    for y in arr2:
      if y not in target_counts or target_counts[y]==0:
        return False
      target_counts[y]-=1
    return True

# Test Cases
print(check_squared_frequency([1,2,3], [1,4,9]))          # True
print(check_squared_frequency([1,2,2], [1,4,4]))          # True
print(check_squared_frequency([1,2,3], [1,4,16]))         # False
print(check_squared_frequency([2,2,3], [4,9,4]))          # True
print(check_squared_frequency([2,2,3], [4,9,9]))          # False
```

# 3. Check Cubed Frequency

```python
def check_cubed_frequency(arr1, arr2):

    if len(arr1) != len(arr2):
        return False

    target_counts = {}

    for x in arr1:
        cube = x ** 3
        target_counts[cube] = target_counts.get(cube, 0) + 1

    for y in arr2:

        if y not in target_counts or target_counts[y] == 0:
            return False

        target_counts[y] -= 1

    return True


# Test Cases
print(check_cubed_frequency([1,2,3], [1,8,27]))      # True
print(check_cubed_frequency([2,3], [8,27]))          # True
print(check_cubed_frequency([2,3], [8,9]))           # False
print(check_cubed_frequency([2,2], [8,8]))           # True
print(check_cubed_frequency([2,2], [8,27]))          # False
```

```python
def top_frequent(arr):

    count = {}

    for num in arr:
        count[num] = count.get(num, 0) + 1

    print("Frequency:", count)

    return max(count, key=count.get)


# Test
arr = [1,1,1,2,2,3]

print(top_frequent(arr))
```

```python
def top_k_frequent(arr, k):

    count = {}

    # Step 1: Count frequency
    for num in arr:
        count[num] = count.get(num, 0) + 1

    print("Frequency:", count)

    # Step 2: Sort by frequency
    sorted_items = sorted(count.items(), key=lambda x: x[1], reverse=True)

    print("Sorted:", sorted_items)

    # Step 3: Return first k elements
    result = []

    for i in range(k):
        result.append(sorted_items[i][0])

    return result


# Test Cases
print(top_k_frequent([1,1,1,2,2,3],2))
print(top_k_frequent([4,4,4,4,5,5,6],2))
print(top_k_frequent([7,7,8,8,8,9],1))
```

# Check Triple Frequency

```python
def check_tripled_frequency(arr1, arr2):

    if len(arr1) != len(arr2):
        return False

    target_counts = {}

    for x in arr1:
        triple = x * 3
        target_counts[triple] = target_counts.get(triple, 0) + 1

    for y in arr2:

        if y not in target_counts or target_counts[y] == 0:
            return False

        target_counts[y] -= 1

    return True


# Test Cases
print(check_tripled_frequency([1,2,3],[3,6,9]))      # True
print(check_tripled_frequency([1,2],[3,5]))          # False
print(check_tripled_frequency([2,2],[6,6]))          # True
```

# Check Half Frequency

```python
def check_half_frequency(arr1, arr2):

    if len(arr1) != len(arr2):
        return False

    target_counts = {}

    for x in arr1:
        half = x / 2
        target_counts[half] = target_counts.get(half,0)+1

    for y in arr2:

        if y not in target_counts or target_counts[y]==0:
            return False

        target_counts[y]-=1

    return True


# Test Cases
print(check_half_frequency([2,4,6],[1,2,3]))      # True
print(check_half_frequency([2,4],[1,3]))          # False
```

# Check Absolute Value Frequency

```python
def check_absolute_frequency(arr1, arr2):

    if len(arr1)!=len(arr2):
        return False

    target_counts={}

    for x in arr1:
        value=abs(x)
        target_counts[value]=target_counts.get(value,0)+1

    for y in arr2:

        if y not in target_counts or target_counts[y]==0:
            return False

        target_counts[y]-=1

    return True


# Test Cases
print(check_absolute_frequency([-1,-2,3],[1,2,3]))      # True
print(check_absolute_frequency([-1,-2,3],[1,3,3]))      # False
```

# Check Character Frequency (Anagram)

```python
def check_anagram(s1,s2):

    if len(s1)!=len(s2):
        return False

    counts={}

    for ch in s1:
        counts[ch]=counts.get(ch,0)+1

    for ch in s2:

        if ch not in counts or counts[ch]==0:
            return False

        counts[ch]-=1

    return True


# Test Cases
print(check_anagram("listen","silent"))      # True
print(check_anagram("hello","world"))        # False
print(check_anagram("race","care"))          # True
```
