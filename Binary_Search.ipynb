## Binary Search Explained

Binary Search is an efficient algorithm for finding an item from a **sorted list of items**. It works by repeatedly dividing in half the portion of the list that could contain the item, until you've narrowed down the possible locations to just one. This makes it significantly faster than a linear search, especially for large datasets.

### Prerequisites

1.  **Sorted Data**: The most crucial prerequisite for Binary Search is that the array or list must be sorted in ascending or descending order.

### How it Works (Iterative Approach)

1.  **Initialization**: Define `low` (start index) and `high` (end index) of the search space.
2.  **Looping**: While `low` is less than or equal to `high`:
    a.  **Midpoint Calculation**: Calculate the middle index `mid = (low + high) // 2`.
    b.  **Comparison**: Compare the `target` value with the element at `mid`:
        *   If `target == arr[mid]`, the item is found. Return `mid`.
        *   If `target < arr[mid]`, the target must be in the left half. Update `high = mid - 1`.
        *   If `target > arr[mid]`, the target must be in the right half. Update `low = mid + 1`.
3.  **Not Found**: If the loop finishes and the item is not found, return -1 (or raise an error, depending on the implementation).

### Time Complexity

*   **Best Case**: O(1) - when the target element is the middle element.
*   **Average Case**: O(log n)
*   **Worst Case**: O(log n) - where 'n' is the number of elements in the array. This logarithmic complexity makes it very fast for large datasets.

### Space Complexity

*   **O(1)** - for the iterative approach, as it uses a constant amount of extra space.
*   **O(log n)** - for the recursive approach, due to the function call stack.

```python
def binary_search(arr, target):
    """
    Performs a binary search on a sorted list to find the target element.

    Args:
        arr (list): A sorted list of elements.
        target: The element to search for.

    Returns:
        int: The index of the target element if found, otherwise -1.
    """
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2  # Calculate the middle index

        # Check if target is present at mid
        if arr[mid] == target:
            return mid
        # If target is greater, ignore left half
        elif arr[mid] < target:
            low = mid + 1
        # If target is smaller, ignore right half
        else:
            high = mid - 1

    return -1  # Target not found

# Example Usage:
my_list = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]

target1 = 23
result1 = binary_search(my_list, target1)
if result1 != -1:
    print(f"Element {target1} is present at index {result1}")
else:
    print(f"Element {target1} is not present in the list")

target2 = 72
result2 = binary_search(my_list, target2)
if result2 != -1:
    print(f"Element {target2} is present at index {result2}")
else:
    print(f"Element {target2} is not present in the list")

target3 = 100
result3 = binary_search(my_list, target3)
if result3 != -1:
    print(f"Element {target3} is present at index {result3}")
else:
    print(f"Element {target3} is not present in the list")
```

#Program 1. Search an Element

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1


arr = [10, 20, 30, 40, 50, 60]

result = binary_search(arr, 40)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```

#Square Root using Binary Search

```python
def square_root(n):

    low=1
    high=n
    ans=0

    while low<=high:

        mid=(low+high)//2

        if mid*mid<=n:
            ans=mid
            low=mid+1

        else:
            high=mid-1

    return ans

print(square_root(25))
```

#First Occurrence

```python
def first_occurrence(arr, target):
    low = 0
    high = len(arr) - 1
    ans = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            ans = mid
            high = mid - 1

        elif arr[mid] < target:
            low = mid + 1

        else:
            high = mid - 1

    return ans

arr = [2,4,4,4,6,8]
print(first_occurrence(arr,4))
```

#Last Occurrence

```python
def last_occurrence(arr,target):

    low = 0
    high = len(arr)-1
    ans = -1

    while low <= high:

        mid = (low+high)//2

        if arr[mid]==target:
            ans = mid
            low = mid+1

        elif arr[mid]<target:
            low = mid+1

        else:
            high = mid-1

    return ans

print(last_occurrence([2,4,4,4,6],4))
```
